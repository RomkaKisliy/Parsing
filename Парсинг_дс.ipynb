```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import requests
```

```python
with open(r"C:\Users\22259910\Downloads\Купить залоги Сбербанка и других банков, активы на продажу.html", 'r', encoding = 'utf-8') as file:
    html_content = file.read()
    
soup = bs(html_content, 'lxml')
print(soup.prettify())
```

```python
asset = soup.find_all("div", class_="asset-title")
```

```python
links = []
for a in asset:
    a_tag = a.find("a", class_="link")
    if a_tag and "href" in a_tag.attrs:
        links.append(a_tag["href"])
```

```python
l = links[0]
response = requests.get(l)

if response.status_code == 200:
    html_content = response.text
    print(html_content)
else:
    print("Ошибка", response.status_code)
```
